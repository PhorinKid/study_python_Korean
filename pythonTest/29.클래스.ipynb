````python
class Test:
    def __init__(self): # 생성자 함수
        print('init call')
        print('init :', id(self))
        self.a = 0 # 맴버 데이터 / self. -> 간접연산자
        self.b = 0

    def setData(self, x, y): # 맴버 함수
        self.a = x
        self.b = y
    
    def show(self): # 맴버 함수
        print(self.a, self.b)

obj = Test() # Test 객체생성 (Test 클래스의 맴버테이터 메모리 할당 완료), 래퍼런스, 변수
# obj.__init__(obj) <- 컴파일러가 추가
print('obj :', id(obj))
obj.setData(10,20) # obj.setData(obj, 10, 20)
obj.show() # obj.show(obj)

obj1 = Test()
# obj.__init__(obj)
print('obj1 :', id(obj1))
obj1.setData(11,22) # obj.setData(obj, 11, 22)
obj1.show() # obj1.show(obj1)
````
