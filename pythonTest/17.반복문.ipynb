#### 01. while

```python
a = 1
while a<=3 :
    print('a =', a)
    a += 1
print('hello')
```

```python
a = 1
while a<=10 :
    print('a =', a)
    a += 1
print('끝')
```

```python
a = 1
while a<=10 :
    if a%2==0 :
        print('a =', a)
    a += 1
print('끝')
```

#### 02. for

요소 in 복합데이터타입
for 변수 in 복합데이터타입

```python
mylist = [10,20,30,40,50]
for n in mylist : # n = mylist[0 ~ -1]
    print(n)
```

```python
mys = 'abcd'
for n in mys : # n = mys[0 ~ -1]
    print(n)
```

```python
myt = (11,22,33,44)
for n in myt : # n = myt[0 ~ -1]
    print(n)
```

```python
myd = {'aa':11, 'bb':22, 'cc':33}
for n in myd : # myd.keys() = ['aa','bb','cc']
    print(n)
```

```python
myd = {'aa':11, 'bb':22, 'cc':33}
for n in myd.values() : # myd.values() = [11,22,33]
    print(n)
```

```python
myd = {'aa':11, 'bb':22, 'cc':33}
for n in myd.items() : # myd.items() = [('aa', 11),('bb', 22),('cc', 33)]
    print(n)
```

```python
myd = {'aa':11, 'bb':22, 'cc':33}
for i,j in myd.items() : # myd.items() = [('aa', 11),('bb', 22),('cc', 33)]
    print(i ,':', j)
```
