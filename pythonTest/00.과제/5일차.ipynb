```python
# 2. 아래코드가 동작하도록 차를 상속받은 자전차 클래스를 만드시오.

class 차:
    def __init__(self, 바퀴, 가격):
        self.바퀴 = 바퀴
        self.가격 = 가격

class 자전차(차):
    def __init__(self, 바퀴, 가격, 구동계):
        super().__init__(바퀴, 가격)
        self.구동계 = 구동계

bicycle = 자전차(2, 100, "시마노")
print(bicycle.바퀴) # 2
print(bicycle.가격) # 100
print(bicycle.구동계) # 시마노
```

```python
# 03. 아래 코드가 동작하도록 Book 클래스를 완성하시요.

class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def __str__(self):
        return f'책제목 : {self.title} / 저자 : {self.author}'

book = Book("데미안", "헤르만 헤세")
print(book) # 책제목:데미안 저자: 헤르만 헤세
```

```python
# 04. 다음이 동작하도록 클래스를 완성하시오.

class Test:
    def __init__(self, a,b):
        self.a = a
        self.b = b
    def setData(self, x,y):
        self.a = x
        self.b = y
    def show(self):
        print( self.a, self.b)       

    def __add__(self, other):
        return (self.a + other.a, self.b + other.b)

obj1 = Test(10,20)
obj2 = Test(100,200)
rst = obj1 + obj2
print(rst) # (110,220)

```

```python
# 05. 두 학생의 점수가 같으면 == 가 True가 되도록, 한 학생의 점수가 더 작으면 < 가 True 가 되도록 구현하세요.
class Student:
    def __init__(self, name, score):
        self.name = name
        self.score = score

    def __eq__(self, other):
        return self.score==other.score

    def __lt__(self, other):
        return self.score<other.score

# __eq__, __lt__ 구현
s1 = Student("민수", 85)
s2 = Student("지연", 90)
s3 = Student("철수", 85)
print(s1 == s3) # True
print(s1 < s2) # True
```
