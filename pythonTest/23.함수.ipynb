#### 01. 디폴트 아규먼트

```python
def fn(a=1,b=2) :
    print("a =", a)
    print("b =", b)
```

```python
fn(10,20)
```

```python
fn(100)
```

```python
fn()
```

```python
fn(b=20) # 명시적 호출(아규먼트)
```

#### 02. 여러개 리턴

```python
def fn1() :
    return 10, 20
```

```python
rst = fn1() # 패킹
print(rst)
```

```python
a, b = fn1()
print(a,b)
```

```python
def shape(r, h, w) :
    return 3.14*r**2, h*w
```

```python
rst = shape(3,10,20)
print(rst)
print(rst[0])
print(rst[1])
```

#### 03.가변인자 : 함수 호출시 아규먼트 갯수 제한 없음

```python
def fn3(*args) :
    print(args)
```

```python
fn3(10,20,30,40,50)
fn3("aa","bb","cc")
```

```python
def circles(*rs) :
    for r in rs :
        print(3.14*r**2)
```

```python
circles(1,2,3,4,5)
```

```python
def fn4(*args, a=0):
    print('args =', args)
    print('a =', a)
```

```python
fn4(1,2,3,4,5)
```

```python
fn4(1,2,3,4, a=5)
```

#### 04. 타입 명시 함수 : 3.5버전 이후 부터
def fn5(a:인자타입, b:인자타입)->반환타입:

```python
def fn5(a:int, b:int)->int:
    return a+b
```

```python
rst = fn5(10,20)
print(rst)
```

```python
def hap(a:object, b:object)->object: # 생략하면 object -> 타입의 제한이 없다.
    return a+b
```

```python
rst = hap(10.1, 20.2)
print(round(rst,1))
```
