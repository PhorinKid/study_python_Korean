#### 01. call by reference

```python
my = [10,20,30]

def fn(aa) :
    aa[0] = 100
    print(aa)

fn( my )
print(my)
```

#### 02. return

```python
def fn() :
    mylist = [10,20,30]
    return mylist

rst = fn()
print(rst)
```

```python
mylist = [10,20,30]
aa = mylist #shallow copy
print(id(aa), id(mylist))
aa[0] = 100
print(mylist)
```

#### 03. deep copy

```python
mylist = [10,20,30]
# aa = [10,20,30]
aa = mylist.copy() # deep copy
print(id(aa), id(mylist))
aa[0] = 100
print(mylist)
```

#### 04. reference count (참조계수 기법) : 파이썬 메모리 관리

```python
mylist = [10,20,30]
mylist = 'abc' # 리스트 객체는 소멸
print(mylist)
```

```python
mylist = [10,20,30]
aa = mylist # 리스트에 대한 reference count
mylist = 'abc'
print(mylist)
print(aa)
```
