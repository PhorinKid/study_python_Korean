### 데이터를 필터링하는 함수

````python
def myfilter1(data) :
    return [n for n in data if n>=30]

def myfilter2(data) :
    return [n for n in data if n<=20]

def myfilter3(data) :
    return [n for n in data if n==20 or n==30]
````

````python
mylist = [10,20,30,40,50]
rst = myfilter1(mylist)
print(rst)
rst = myfilter2(mylist)
print(rst)
rst = myfilter3(mylist)
print(rst)
````

#### 01. 일급 함수 기능 : open cloesd 기법

````python
def myfilter(data, key) :
    return [n for n in data if key(n)]

def fn1(v) :
    return v>=30

def fn2(v) :
    return v<=20

def fn3(v) :
    return v==20 or v==40
````

````python
mylist = [10,20,30,40,50]
rst = myfilter(mylist, fn2)
print(rst)
````

````python
def myfilterl(data, key) :
    return [n for n in data if key(n)]

mylist = [10,20,30,40,50]
rst = myfilterl(mylist, lambda v:v>=30)
print(rst)
rst = myfilterl(mylist, lambda v:v<=20)
print(rst)
````

````python
 # mymap 함수를 만드시오.

def mymap(data, key) :
    return [key(n) for n in data]

mylist = [10,20,30,40,50]

rst = mymap(mylist, lambda v:v+5)
print(rst)
````
