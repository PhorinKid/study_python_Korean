#### 01. 함수 : 리턴, 인자(아규먼트), 할당, 함수 내에 함수를 선언(데코레이터)

##### 01_1. 할당

```python
def fn() :
    print('fn call')

aa = fn
print(id(aa), id(fn), id(fn())) # fn에 할당된 주소
print(aa)
aa()
```

#### 아규먼트

```python
def fn() :
    print('fn call')

def afn(aa) :
    print('aa', id(aa))
    aa()

print('fn', id(fn))
afn( fn )
```
