```python
mylist = [10,20,30,40,50]
result = [n*10 for n in mylist]
print(result)
```

```python
# 세금 3.3% 제한 실수령액 리스트를 구하시오

salary = [1000,2000,3000,4000,5000]
result = [n*(1-0.033) for n in salary]
print(result)
```

```python
rst = [n for n in range(1,6) if n>=3]
print(rst)
```

```python
# 1 ~ 10 까지 짝수 리스트
rst = [n for n in range(1,11) if n%2==0]
print(rst)
```

```python
# '이' 글자가 있는 이름에 대한 리스트를 구하시오

name = ['김철수','홍길동','이순신','이황','이이']
rst = [n for n in name if '이' in n]
print(rst)
```
