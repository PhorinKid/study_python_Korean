### 문자열, 리스트, 튜플, 세트, 딕셔너리

#### 01. 문자열 타입

```python
s1 = "abc\
def"
s2 = 'abc' \
'def'
s3 = """abc
def"""
s4 = '''abc
    def'''
print(s1)
print(type(s1))
print(s2)
print(type(s2))
print(s3)
print(type(s3))
print(s4)
print(type(s4))
```

```python
s1 = "abc\ndef"
s2 = 'abc\tdef'
print(s1)
print(s2)
```

```python
s5 = 'hello "Tom"'
print(s5)
```

#### 02. 리스트 타입 [ ] : 변경 가능 (mutable) - 내용 추가, 수정, 삭제 가능

```python
myList = [10, 20, 30, 40, 50]
myList1 = [1.1, 2.2, 3.3]
myList2 = ['aa', 'bb', 'cc']
print(myList)
print(type(myList))
print(myList1)
print(type(myList1))
print(myList2)
print(type(myList2))
```

#### 03. 튜플 타입 ( ) : 변경불가 (immutable) - 내용 추가, 수정, 삭제 불가

```python
t = (10, 20, 30, 40, 50)
print(t)
print(type(t))
```

#### 04. set{ } : 변경가능(mutable), 순서없음, 집합연산

```python
s = {10, 20, 30, 40, 50}
print(s)
print(type(s))
```

#### 05. 딕셔너리 { 키 : 값 } : 변경가능(mutable), 순서없음

```python
d = { 'aa':10, 'bb':20, 'cc':30 }
print(d)
print(type(d))
```
