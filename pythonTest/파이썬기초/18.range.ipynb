``` text
리스트를 만들어 주는 함수
- range (시작값, 끝값, 증가치)
- 시작값 <= 리스트 < 끝값

````python
a = list(range(1,10,2)) # 1 <= 리스트 < 10 , 증가치 2
print(a)
print(type(a))
````

````python
a = list(range(10)) # 시작이 0 <= 리스트 < 10 , 증가치 1
print(a)
print(type(a))
````

````python
for n in range(1,5) :
    print(n)
````

``` text
- break
- continue

````python
for n in range(1,11) :
    if n == 4 :
        break # 멈추기
    print(n)
````

````python
for n in range(1,11) :
    if n == 4 :
        continue # 건너 뛰기
    print(n)
````

````python
# 1 ~ 10 까지의 정수 중 3 의 배수를 제외하고 출력 (continue 사용)

for n in range(1, 11) :
    if n%3==0 :
        continue
    print(n)
````
