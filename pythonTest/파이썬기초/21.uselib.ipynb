```python
import mylib # mylib.py 에 있는 함수를 사용 (선언부)

rst = mylib.hap(10,20) # mylib 에 있는 hap 함수 호출
print(rst)
rst = mylib.gop(10,20) # mylib 에 있는 gop 함수 호출
print(rst)
```

```python
import mylib as m # as m ==> 약어 설정

rst = m.hap(10,20) # mylib 에 있는 hap 함수 호출
print(rst)
rst = m.gop(10,20) # mylib 에 있는 gop 함수 호출
print(rst)
```

```python
from mylib import hap, gop # mylib.py 에 있는 hap, gop 함수 사용 선언

rst = hap(10,20) # mylib 에 있는 hap 함수 호출 / mylib 작성 불필요
print(rst)
rst = gop(10,20) # mylib 에 있는 gop 함수 호출 / mylib 작성 불필요
print(rst)
```

```python
# circle.py 에 원과 관련된 함수를 모아높은 라이브러리 만들기 (원의 면적, 원기둥의 부피, 원뿔의 부피)
# 라이브러리 import 해서 호출 후 결과 

import circle as c

rst = c.area(5)
print(rst)
rst = c.cyl(5,10)
print(rst)
rst = c.cone(5,10)
print(round(rst,2))
```
