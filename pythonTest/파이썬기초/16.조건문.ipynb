```python
scr = int(input('점수 : '))

if scr>100 :
    print('점수는 100점 이하')
elif 100>=scr>=90 :
    print('A')
elif 90>scr>=80 :
    print('B')
elif 80>scr>=70 :
    print('C')
elif 70>scr>=60 :
    print('D')
else :
    print('F')
```
