### 인덱싱, 슬라이싱
#### 복합데이터 타입에서 특정 데이터 추출시 사용
``` text
- 순서 있는 데이터 타입 : 문자열, 리스트, 튜플
- 순서 없는 데이터 타입 : 세트, 딕셔러니 (인덱싱, 슬라이싱 적용 불가능)

#### 01. 인덱싱 : [인덱스번호]

``` text
- 0, 1, 2, ...
- ..., -3, -2, -1

````python
#      0   1   2   3   4
#     -5  -4  -3  -2  -1
my = [10, 20, 30, 40, 50]
print(my[0])
print(my[-5])
my[3]
````

````python
t = (0, 2, 4 ,6)
print(t[0])
print(t[-1])
````

````python
s = 'abcde'
print(s[2])
print(s[-3])
````

````python
s = {10, 20, 30}
print(s[0]) # error
````

````python
d = {'aa':10, 'bb':20, 'cc':30}
# print(d[0]) # error
print(d['aa'])
````

#### 02. 슬라이싱

``` text
- [시작인덱스:끝인덱스:증가치]
- 시작인덱스 <= 인덱싱 < 끝인덱스

````python
my = [10, 20, 30, 40, 50]
print(my[1:4:1]) # 1 <= index < 4 ==> 1, 2, 3
print(my[1:4:2]) # 1 <= index < 4 ==> 1, 3
print(my[0:5:2]) # 0 <= index < 5 ==> 0, 2, 4
````

````python
my[1:4] # 증가치 생략 시 1씩 증가 ==> my[1:4:1]
````

````python
my[:4] # 시작인덱스 생략 시 0시작 ==> my[0:4:1]
````

````python
my[1:] # 끝인덱스 생략 시 끝까지 ==> my[1:5:1]
````

````python
my[:-1] # 마지막 제외 슬라이싱 ==> my[0:4:1] / my[0:-1:1]
````

````python
my[-3:] # -3 <= index ==> -3, -2, -1 / 마지막 3개 데이터 출력
````

````python
my[-3::-1] # -3 <= index ==> -1씩증가 / -3, -4, -5
````

````python
my[-1::-1] # 역 리스트 / -1, -2, -3, ...
````

````python
my[::-1] # 증가치가 '-'(감소)일 경우 시작은 자동 -1
````

````python
my[::-2] # 증가치 -2
````
