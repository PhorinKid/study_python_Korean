1. 1 cm 가 0.393701 inch 인 경우 7 cm 가 몇 inch 인지 구하시오.
2. 반지름이 5 이고 높이가 10 인 원기중의 부피를 구하시오. pi = 3.14
3. 폭이 10 이고 높이가 5 인 사각형의 면적을 구하시오.
4. 정수가 5 인경우 짝수인지 홀수인지 구하시오.
5. fruit = ["사과", "포도", "홍시"] 에서 "홍시" 요소가 있는지 확인하시오. 결과 값 '있다' 또는 '없다'

```python
# 01
cm = 7
inch = 0.393701
rst = cm * inch
print(rst)
```

```python
# 02
r, h, pi = 5, 10, 3.14
srt = r**2 * pi * h
print(srt)
```

```python
# 03
w, h = 10, 5
area = w * h
print(area)
```

```python
# 04
a = 5
srt = '짝수' if a%2==0 else '홀수'
print(srt)
```

```python
# 05
fruit = ["사과", "포도", "홍시"]
srt = '있다' if '홍시' in fruit else '없다'
print(srt)
```
