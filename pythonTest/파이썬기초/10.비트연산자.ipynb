비트연산자 (2진법)

````python
a = 3
rst = a & 1
print(rst)
a = 3
rst = a & 4
print(rst)
````

```text
비트 and(&) 모두 1인경우 1 / 하나라도 0 이면 0 
# 논리곱
    00000011
&   00000001
============
    00000001 = 1

````python
a = 3
rst = a | 1
print(rst)
````

```text
비트 or(|) 하나라도 1 인경우 1 / 모두 0 이면 0 
# 논리합
    00000011
&   00000001
============
    00000011 = 3

````python
a = 3
rst = a << 1
print(rst)
````

```text
<< (이동 수)
    00000011
<< 1
============
    00000110 ==> 6

````python
a = 3
rst = a >> 1
print(rst)
````

```text
>> (이동 수)
    00000011
>> 1
============
    00000001 ==> 1
