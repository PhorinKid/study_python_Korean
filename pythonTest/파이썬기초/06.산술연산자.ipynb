산술연산자 우선순위
1. ** (지수연산자)
2. \* (곱하기), / (나누기), % (나머지), // (몫)
3. \+ (더하기), -(뺴기)

#### 01. 지수연산자

```python
print(2**2)
print(2**3)
```

```python
a = 5
rst =  a**2
print(rst)
```

#### 02. 곱하기

```python
a = 5
rst =  a*2
print(rst)
```

#### 03. 나누기

```python
a = 5
rst =  a/2
print(rst)
```

#### 04. 나머지

```python
a = 5
rst =  a%2
print(rst)
```

#### 05. 몫

```python
a = 5
rst =  a//2
print(rst)
```

#### 06. 더하기

```python
a = 5
rst =  a+2
print(rst)
```

#### 07. 빼기

```python
a = 5
rst =  a-2
print(rst)
```

#### 우선순위 연습

```python
rst = 3 + 2**2 * 2 - 1
print(rst)
```
