#### 01. 정수 타입


```python
a = 10
b = 123456789
print(a)
print(b)
print(type(b))
```

#### 02. 실수 타입

```python
f0 = 3.14
f1 = 3.141592e3
f2 = 3.141592e-3
print(f0)
print(type(f0))
print(f1)
print(type(f1))
print(f2)
print(type(f2))
```

#### 03. bool 타입

```python
bb = True
CC = False
print(bb)
print(CC)
print(type(bb))
```
