관계연산자 (관계연산자는 산술연산자 보다 우선순위가 낮다) : 연산의 결과는 bool 타입
``` text
- > (크다)
- >= (크거나 같다)
- < (작다)
- <= (작거나 같다)
- == (같다)
- != (같지 않다)

````python
a = 5
rst = a > 0
print(rst)
````

````python
a = 5
rst = a >= 5
print(rst)
````

````python
a = 5
rst = a < 5
print(rst)
````

````python
a = 5
rst = a <= 4
print(rst)
````

````python
a = 5
rst = a == 5
print(rst)
````

````python
a = 5
rst = a != 4
print(rst)
````

````python
a = 3
rst = 0 <= a <= 10 # 파이썬에서만 가능

print(rst)
````
