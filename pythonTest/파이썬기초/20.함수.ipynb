```python
def hap(a,b) :
    rst = a+b
    return rst
```

```python
def gop(a,b) :
    rst = a*b
    return rst
```

```python
def minus(a,b):
    rst = a-b
    return rst
```

```python
def circle(r) :
    rst = r**2*3.14
    return rst
```

```python
rst = hap(10,20)
print(rst)
```

```python
rst = gop(10,20)
print(rst)
```

```python
rst = minus(10,20)
print(rst)
```

```python
rst = circle(10)
print(rst)
```

```python
# 점수를 인로 받아 70 이상이면 합격인 경우, 합격 여부를 반환하는 함수

def isPass(score) :
    # if score>=70 :
    #     return '합격'
    # else :
    #     return '불합격'
    return '합격' if score>=70 else '불합격'

rst = isPass(70)
print(rst)
```

```python
# 1부터 n 까지 리스트를 만드는 함수

def intList(num) :
    return [n for n in range(1, num+1)]

print(intList(10))
```

```python
# 01. cm 를 인자로 받아 inch 를 반환하는 함수

def cmInch(cm) :
    inch = 0.3937
    rst = cm * inch
    return rst

print(cmInch(10))
```

```python
# 02. 반지름과 높이를 인자로 받아 원기중의 부피를 반환하는 함수

def cyl(r,h) :
    v = r**2*3.14*h
    return round(v,2)

print(cyl(5,10))
```
