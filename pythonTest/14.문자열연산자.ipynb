```text
문자열연산자 : + , * , %
- 문자열 연결 : +
- 문자열 반복 : *
- 포맷 문자열 : %

````python
s = 'abc'
s = s + 'def'
print(s)
````

````python
s = 'abc'
s = s * 3
print(s)
````

````python
a = 10
b = 3.14
c = 'korea'
# s = 'a=%d b=%f c=%s' %(a, b, c) ==> 구세대 방법
s = f'a={a:10} b={b:<10.1f} c={c:>10}' # 3.5 버전 이후 포맷
print(s)
````
