#### 01. 문자열 맴버함수

````python
s = 'abcdeF'
print(s.upper()) # 소문자를 대문자로
````

````python
s = '     abc     '
print(s)
print(s.strip()) # 문자열 좌우의 화이트스페이스(' '공백, \n, \t, \r) 제거
````

````python
s1 = '###abc---'
print(s1.strip('#-')) # 좌우의 '#' 혹은 '-' 제거
````

````python
s2 = 'abc def gfi'
print(s2.split()) # 화이트스페이스 기준으로 분리 - 리스트
````

````python
s3 = '000-111-222'
print(s3.split('-')) # '-'를 기준으로 분리
````

````python
s4 = 'i love korea'
print(s4.replace('love', 'like'))
````

#### 02. 리스트 맴버함수

````python
myList = [10, 20, 30]
myList.append(40) # 맨 뒤에 삽입
print(myList)
myList.append(50)
print(myList)

myList.insert(1,100) # 1번 자리에 삽입
print(myList)
````

````python
myList = [10, 20, 30]
myList[0] = 100 # 0 위치의 값을 수정
print(myList)
````

````python
myList = [10, 20, 30]
myList.remove(20) # 삭제
print(myList)
````

````python
my1 = ['aa', 'bb', 'cc']
print(' '.join(my1)) # 리스트를 공백으로 구분하여 문자열
print('-'.join(my1)) # 리스트를 '-'으로 구분하여 문자열
````

#### 03. 튜플 맴버함수

````python
t = (10, 20, 30, 20)
print(t.count(20)) # 해당 데이터의 갯수
print(t.index(20,2)) # 2번째 20의 위치
````

#### 04. 패킹

````python
a = 10; b = 20; c = 30
print(a, b, c)
````

````python
a, b, c = 10, 20, 30
print(a, b, c)
````

````python
d = 11, 22, 33, 11 # 패킹 : 자동으로 튜플로 변환
print(d)
````

````python
e, f, g = (100, 200, 300) # 언패킹 : unpacking - 리스트, 튜플 등을 각각의 변수로 분리
print(e, f, g)
````

````python
d = 11, 22, 33, 11
print(d)
print(*d) # *사용시 자동으로 언패킹 (선언 X)
````

#### 05. 딕셔너리 맴버함수

````python
d = {'aa':10, 'bb':20, 'cc':30}
print(d.keys()) # 키만 출력
print(d.values()) # 값만 출력
print(d.items()) # 각 키와 값을 튜플 묶은 후 리스트에 저장
````

````python
d = {'aa':10, 'bb':20, 'cc':30}
print(d['aa'])
print(d.get('aa')) # 키의 값을 출력
print(d.get('ff', 100)) # 키의 값이 없으면 기본값 출력(추가 x)
print(d)
````

````python
d = {'aa':10, 'bb':20, 'cc':30}
d.pop('aa') # 'aa' 삭제
print(d)
````

````python
d = {'aa':10, 'bb':20, 'cc':30}
d['aa'] = 100 # 키가 있으면 수정
d['dd'] = 200 # 키가 없으면 추가
print(d)
````

#### 06. set 맴버함수

````python
s = {10, 20, 30, 40, 50}
s.add(60) # 추가
s.remove(20) # 삭제
print(s)
````

````python
s = {10, 20, 30, 40, 50, 20, 20, 30} # 중복데이터 자동 제거
print(s)
````

````python
s1 = {10, 20, 30}
s2 = {20, 30, 40, 50}
print(s1.intersection(s2)) # 교집합
print(s1 & s2) # 교집합
print(s1.union(s2)) # 합집합
print(s1 | s2) # 합집합
print(s1.difference(s2)) # 차집합
print(s1 - s2) # 차집합
````
