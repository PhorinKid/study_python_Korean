```python
class Test:
    st = 100 # static 맴버 (객체 생성X), 클래스 변수
    def __init__(self):
        self.a = 0 # instance 변수(반드시 객체 생성 필요)
        self.b = 0

    def setData(self, x, y):
        self.a = x
        self.b = y
    
    def show(self):
        print(self.a, self.b)

print(Test.st)
a = Test()
b = Test()
print(a.st)
print(b.st)

print(type(Test()))
```
