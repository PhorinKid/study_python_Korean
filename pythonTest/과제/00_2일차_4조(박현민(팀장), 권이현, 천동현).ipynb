```python
# 06. 점수 구간에 해당하는 학점이 아래와 같이 정의되어 있다. 사용자로부터 score 를 입력받아 학점을 출력하라.

score6 = int(input('점수 : '))
if score6>100 :
    print('점수는 100 점 이하')
elif 100>=score6>80 :
    print('A')
elif 80>=score6>60 :
    print('B')
elif 60>=score6>40 :
    print('C')
elif 40>=score6>20 :
    print('D')
elif 20>=score6>=0 :
    print('E')
else :
    print('점수는 0 점 이상')
```

```python
# 11. 키와 몸무게를 입력받아 비만도를 구하고 결과를 출력하시요.

신장 = float(input('신장 : (신장은 100 이상입니다.)'))
몸무게 = float(input('몸무게 : '))

표준체중 = (신장-100) * 0.85
비만도 = 몸무게/표준체중*100

print(비만도)

if 비만도<=90 :
    print('저체중')
elif 90<비만도<=110 :
    print('정상')
elif 110<비만도<=120 :
    print('과제중')
else :
    print('비만')
```

```python
# 17. 상품가격과 지불액을 입력받아 거스름돈을 최소화 하도록 잔돈의 갯수를 구하시요.

price1 = int(input('가격 : '))
money1 = int(input('잔돈 : '))
s = (500, 100, 50, 10)

money1 = money1-price1
print('잔돈 :', money1)
for n in s :
    c1 = money1//n
    money1 %= n
    print(n, '원 :', c1)
```

```python
# 19. 사용자로부터 달러, 엔, 유로, 또는 위안 금액을 입력받은 후 이를 원으로 변환하는 프로그램을 작성하라. 
# 각 통화별 환율은 다음과 같다. 사용자는 100 달러, 1000 엔, 13 유로, 100 위안과 같이 금액과 통화명 사이에 공백을 넣어 입력한다고 가정한다.

money2 = input('금액 입력 :')
m1, c1 = money2.split()
ch1 = {'달러':1167, '엔':1.096, '유로':1268, '위안':171}
rst = ch1[c1] * int(m1)

print(f'{rst:.2f} 원')
```
