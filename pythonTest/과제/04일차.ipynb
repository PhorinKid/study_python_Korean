```python
# 01. 아래 처럼 동작 하도록myprint함수를 구현하시오.

def myprint(mylist1, key):
    for n in mylist1:
        key(n)
    # [key(n) for n in mylist1]

mylist = [10,20,30]
myprint(mylist, lambda v:print(v))
myprint(mylist, lambda v:print(v,end=' '))
```

```python
# 02. 아래 맴버를 가지는 Stock 클래스를 정의하시오.

class Stock:
    def __init__(self, 종목명, 종목코드, PER, PBR, 배당수익률):
        self.종목명 = 종목명
        self.종목코드 = 종목코드
        self.PER = PER
        self.PBR = PBR
        self.배당수익률 = 배당수익률

    def setData(self, 종목명, 종목코드, PER, PBR, 배당수익률):
        self.종목명 = 종목명
        self.종목코드 = 종목코드
        self.PER = PER
        self.PBR = PBR
        self.배당수익률 = 배당수익률

    def show(self):
        print(self.종목명, self.종목코드, self.PER, self.PBR, self.배당수익률)

삼성 = Stock("삼성전자", "005930", 15.79, 1.33, 2.83)
print(삼성.배당수익률)
삼성.show()
```

```python
# 03. 아래의 표를 참조하여 3종목에 대해 객체를 생성하고 각각 show 함수를 통해 내용을 출력하시오.

class Stock:
    def __init__(self):
        self.종목명 = 0
        self.종목코드 = 0
        self.PER = 0
        self.PBR = 0
        self.배당수익률 = 0

    def setData(self, 종목명, 종목코드, PER, PBR, 배당수익률):
        self.종목명 = 종목명
        self.종목코드 = 종목코드
        self.PER = PER
        self.PBR = PBR
        self.배당수익률 = 배당수익률

    def show(self):
        print(self.종목명, self.종목코드, self.PER, self.PBR, self.배당수익률)

삼성 = Stock()
삼성.setData("삼성전자", "005930", 15.79, 1.33, 2.83)
삼성.show()

현대 = Stock()
현대.setData("현대차", "005380", 8.70, 0.35, 4.27)
현대.show()

LG = Stock()
LG.setData("LG전자", "066570", 317.34, 0.69, 1.37)
LG.show()
```

```python
# 04. Account 클래스를 만든 후 객체를 생성하여 저장된 정보를 출력하는 display_info() 메서드를 이용하여 아래처럼 출력하시오.

class Account:
    def __init__(self):
        self.은행이름 = 0
        self.예금주 = 0
        self.계좌번호 = 0
        self.잔고 = 0
    
    def setDara(self, 은행이름, 예금주, 계좌번호, 잔고):
        self.은행이름 = 은행이름
        self.예금주 = 예금주
        self.계좌번호 = 계좌번호
        self.잔고 = 잔고
    
    def display_info(self):
        print('은행이름 :', self.은행이름)
        print('예금주 :', self.예금주)
        print('계좌번호 :', self.계좌번호)
        print('잔고 :', format(self.잔고, ','),'원')

scB = Account()
scB.setDara('sc은행', '파이썬', '111-11-111111', 10000)
scB.display_info()
```
