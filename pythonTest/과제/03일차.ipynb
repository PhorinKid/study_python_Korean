```python
# 01. 점수를 인자로 받아 학점을 구하는 함수를 만드시오.

def rank1(scr) :
    if 100>=scr>80 :
        return 'A'
    elif 80>=scr>60 :
        return 'B'
    elif 60>=scr>40 :
        return 'C'
    else :
        return 'f'

print(rank1(90))
```

```python
# 02. 키와 몸무게를 인자로 받아 비만도를 구하는 함수를 만드시오.

def abcd2(신장, 몸무게) :
    표준체중 = (신장-100) * 0.85
    비만도 = 몸무게/표준체중*100

    print(비만도)

    if 비만도<=90 :
        return '저체중'
    elif 90<비만도<=110 :
        return '정상'
    elif 110<비만도<=120 :
        return '과제중'
    else :
        return '비만'

print(abcd2(170,65))
```

```python
# 03. 화씨를 인자로 받아 도를 반환하는 함수를 만드시오.

def cToF3(f) :
    return (f-32)*5/9

print(cToF3(212))
```

```python
# 04. 2개의 정수를 인자로 받아 절대값의 합을 반환하는 함수를 만드시오.

def plus4(a:int, b:int) :
    a = -a if a<0 else a
    b = -b if b<0 else b
    return a+b

print(plus4(-3,5))
```

```python
# 05. 3개의 정수를 인자로 받아 가장 큰 값을 반환하는 함수를 만드시오.

def max5(a:int, b:int, c:int) :
    s = [a, b, c]
    for n in s :
        if a < n :
            d = n
    return d

print(max5(5, 10, 2))
```

```python
# 06. 연도를 인자로 받아 윤년 여부를 반환하는 함수를 만드시오.

def yun6(a) :
    year6 = int(a)
    return '윤년이다.' if (year6%4==0 and year6%100!=0) or year6%400==0 else '윤년이 아니다.'

print(yun6(2020))
```

```python
# 07. 연도를 인자로 받아 해당연도의 띠(12지신)를 반환하는 함수를 만드시오.

def god7(a) :
    god = ('신','유','술','해','자','축','인','묘','진','사','오','미')
    return god[a%12]

print(god7(2025))
```

```python
# 08. 상품가격과 지불액을 인자로 받아 거스름돈을 최소화 하도록 잔돈의 갯수를 반환하는 함수를 만드시오.

def coin8(a, b) :
    d = [500, 100, 50, 10]
    m = b - a
    s = []

    for n in d :
        s.append(m//n)
        m %= n
    return f'{d[0]:3}원 : {s[0]:2}개\n{d[1]:3}원 : {s[1]:2}개\n{d[2]:3}원 : {s[2]:2}개\n{d[3]:3}원 : {s[3]:2}개'

print(coin8(1320,2000))
```

```python
# 09. 하나의 정수를 인자로 받아 해당 정수의 약수를 반환하는 함수를 만드시오.

def abcd9(a:int) :
    s = []
    for n in range(1, a+1) :
        if a%n==0 :
            s.append(n)
    return s

print(abcd9(100))
```
