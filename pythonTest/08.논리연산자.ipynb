#### 01. and, or, not

````python
a = 5
rst = a>0 and a==5
print(rst)
rst = a>5 and a==5
print(rst)
````

````python
a = 5
rst = a>5 or a==5
print(rst)
rst = a>5 or a==0
print(rst)
````

````python
a = 5
rst = not a==0
print(rst)
rst = not a==5
print(rst)
````

#### 02. in, not in (요소 in 복합테이터타입)

````python
my = [10,20,30,40,50]
rst = 10 in my
print(rst)
rst = 100 in my
print(rst)
````

````python
rst = 'llo' in 'hello korea'
print(rst)
rst = 'abc' in 'hello korea'
print(rst)
````

````python
myd = {'aa':10, 'bb':20, 'cc':30}
rst = 'aa' in myd # 기본은 myd.keys() ==> ['aa','bb','cc']
print(rst)
rst = 10 in myd
print(rst)
rst = 10 in myd.values() # myd.values() ==> [10,20,30]
print(rst)
````

````python
my = [10,20,30,40,50]
rst = 10 not in my
print(rst)
rst = 100 not in my
print(rst)
````

#### 03. False
```text
- False : 0, [], {}, '' - (비어있는 복합데이터) , None, Nall
- True : 0이 아닌 정수나 실수, [10], {'aa':0}

````python
rst = not 0
print(rst)
rst = not 100
print(rst)
rst = not ''
print(rst)
rst = not 'a'
print(rst)
rst = not []
print(rst)
rst = not [0]
print(rst)
````
